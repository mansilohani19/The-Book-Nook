```python
def age_group(age):
    if pd.isna(age):
        return "Unknown"
    elif age <= 18:
        return "Teen"
    elif age <= 25:
        return "Young_Adult"
    elif age <= 35:
        return "Adult"
    elif age <= 50:
        return "Middle_Age"
    else:
        return "Senior"
```

```python
df_merge['Age_Group'] = df_merge['Age'].apply(age_group)
```

```python
# Unique users aur books nikaalo
unique_users = df_merge['User_ID'].nunique()
unique_books = df_merge['ISBN'].nunique()
total_ratings = len(df_merge)

total_possible = unique_users * unique_books

sparsity = 1 - (total_ratings / total_possible)

print(f"Unique Users: {unique_users}")
print(f"Unique Books: {unique_books}")
print(f"Total Ratings: {total_ratings}")
print(f"Total Possible Ratings: {total_possible}")
print(f"Sparsity: {sparsity:.4f} or {sparsity*100:.2f}%")
```
